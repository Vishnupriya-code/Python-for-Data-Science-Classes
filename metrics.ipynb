```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt

df= pd.read_csv(r"C:\Users\vishn\Downloads\creditcard.csv")[:80_000] #first 80,000 rows
df.head(3)
```

```python
X = df.drop(columns=['Time', 'Amount', 'Class']).values   #data cleaning, independent variables
y = df['Class'].values #it needs to be predicted variables 
f"shapes of X={X.shape} y={y.shape}, #Fraud cases={y.sum()}"    #shows how unbalanced the dataset is 
```

```python
from sklearn.linear_model import LogisticRegression #sigmoid function

mod = LogisticRegression(class_weight = {0 : 1, 1 : 2},max_iter=1000) #dictionaries
mod.fit(X,y).predict(X).sum()
```

```python
from sklearn.model_selection import GridSearchCV #heart of metrics
from sklearn.metrics import precision_score, recall_score, make_scorer 

grid = GridSearchCV(
     estimator=LogisticRegression(max_iter=1000),
     param_grid={'class_weight': [{0 : 1, 1 : v} for v in range(1,4)]},
     #scoring={'precison': make_scorer(precision_score), 'recall': make_scorer(recall_score)},
     #refit='precision',
     return_train_score=True,
     cv=4,
     n_jobs=-1
)
grid.fit(X,y);
```

```python
pd.DataFrame(grid.cv_results_)
```
